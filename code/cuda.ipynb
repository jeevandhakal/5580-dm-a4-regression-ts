```python
from pathlib import Path
import os
import tensorflow as tf

# Define the standard Zorin/Ubuntu CUDA path
cuda_base = Path("/usr/lib/cuda")

if cuda_base.exists():
    # Modern path resolution
    cuda_bin = cuda_base / "bin"
    cuda_lib = cuda_base / "lib64"
    
    # Update Environment using Pythonic string joining
    os.environ["PATH"] = f"{cuda_bin}:{os.environ.get('PATH', '')}"
    os.environ["LD_LIBRARY_PATH"] = f"{cuda_lib}:{os.environ.get('LD_LIBRARY_PATH', '')}"
    
    # Tell TensorFlow where to find the XLA/NVVM data
    os.environ["XLA_FLAGS"] = f"--xla_gpu_cuda_data_dir={cuda_base}"
    
    print(f"‚úÖ CUDA Paths linked from: {cuda_base}")
else:
    print("‚ùå CUDA base path not found. Check if nvidia-cuda-toolkit is installed.")

# Verification
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"üöÄ Success! GPU Detected: {gpus}")
else:
    print("‚ö†Ô∏è GPU not detected. TensorFlow is still defaulting to CPU.")
```
